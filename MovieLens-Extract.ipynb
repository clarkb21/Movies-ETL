```python
#Changing data types
#Check that 'adult' and 'video' are either True or False so they can be changed to boolean
movies_metadata['adult'].value_counts()
```

```python
movies_metadata[~movies_metadata['adult'].isin(["True", "False"])]
```

```python
#Removing bad data and removing "adult" films for the public friendly hackathon event
kaggle_metadata = movies_metadata[movies_metadata['adult'] == 'False'].drop('adult', axis = 'columns')
kaggle_metadata.head()
```

```python
#look at video column data
kaggle_metadata['video'].value_counts()
```

```python
#Converting data type
kaggle_metadata['video'] = kaggle_metadata['video'] == 'True'

kaggle_metadata.dtypes
```

```python
kaggle_metadata['budget'] = kaggle_metadata['budget'].astype(int)
kaggle_metadata['id'] = pd.to_numeric(kaggle_metadata['id'], errors='raise')
kaggle_metadata['popularity'] = pd.to_numeric(kaggle_metadata['popularity'], errors='raise')

#check data types
kaggle_metadata.dtypes
```

```python
#Convert release date to datetime data type
kaggle_metadata['release_date'] = pd.to_datetime(kaggle_metadata['release_date'])

#Check data types again
kaggle_metadata.dtypes
```
